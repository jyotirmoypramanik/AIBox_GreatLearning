```python
# Helper function to display responses as Markdown, horizontally
def display_responses(*args):
    markdown_string = "<table><tr>"
    # Headers
    for arg in args:
        markdown_string += f"<th>System Prompt:<br />{arg['system_prompt']}<br /><br />"
        markdown_string += f"User Prompt:<br />{arg['user_prompt']}</th>"
    markdown_string += "</tr>"
    # Rows
    markdown_string += "<tr>"
    for arg in args:
        markdown_string += f"<td>Response:<br />{arg['response']}</td>"
    markdown_string += "</tr></table>"
    display(Markdown(markdown_string))
```

```python
def display_responses_modified(*args):
    markdown_string = "<table><tr>"
    # Headers
    for arg in args:
        markdown_string += f"<th>System Prompt:<br />{arg['system_prompt']}<br /><br />"
        markdown_string += f"User Prompt:<br />{arg['user_prompt']}</th>"
    markdown_string += "</tr>"
    # Rows
    markdown_string += "<tr>"
    for arg in args:
        markdown_string += f"<td>Response:<br />{arg['response']}<br /><br />"
        # ── Stats bar (only shown if stats key exists) ────────────
        if "stats" in arg:
            s = arg["stats"]
            markdown_string += (
                f"<small>"
                f"⏱ <b>{s['total_time_sec']}s</b> &nbsp;|&nbsp; "
                f"🚀 <b>{s['tokens_per_sec']} tok/s</b> &nbsp;|&nbsp; "
                f"📈 <b>{s['tokens_per_min']} tok/min</b> &nbsp;|&nbsp; "
                f"📥 prompt: <b>{s['prompt_tokens']}</b> &nbsp;|&nbsp; "
                f"📤 output: <b>{s['completion_tokens']}</b> &nbsp;|&nbsp; "
                f"🔢 total: <b>{s['total_tokens']}</b> &nbsp;|&nbsp; "
                f"✅ <b>{s['finish_reason']}</b>"
                f"</small>"
            )
        markdown_string += "</td>"
    markdown_string += "</tr></table>"
    display(Markdown(markdown_string))
```

```python
# Function to interact with the OpenAI API
from enum import Enum
import requests

# ── 1. Basic Enum  (direct mirror of your OpenAIModels pattern) ──
class OllamaModels(str, Enum):
    QWEN3_4B    = "qwen3:4b"       # ~2.6 GB  — fast, multilingual
    GEMMA3_4B   = "gemma3:4b"      # ~3.3 GB  — Google, strong reasoning
    PHI4_MINI   = "phi4-mini"      # ~3.8 GB  — Microsoft, code-focused
    
    def __str__(self) -> str:
        return self.value          # so f"{OllamaModels.QWEN3_4B}" → "qwen3:4b"
#class OpenAIModels(str, Enum):
#    GPT_4O_MINI = "gpt-4o-mini"
#    GPT_41_MINI = "gpt-4.1-mini"
#    GPT_41_NANO = "gpt-4.1-nano"
# MODEL = OpenAIModels.GPT_4O_MINI


MODEL = OllamaModels.QWEN3_4B


def get_completion(system_prompt, user_prompt, model=MODEL):
    """
    Function to get a completion from the OpenAI API.
    Args:
        system_prompt: The system prompt
        user_prompt: The user prompt
        model: The model to use (default is gpt-4.1-mini)
    Returns:
        The completion text
    """
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
            ],
            temperature=0.7,
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}"

print(" Done ... fetching Model ...")
```

```python
# TODO: Add a role-based prompt for portraying Albert Einstein where you see the **********
# baseline_system_prompt = "You are Albert Einstein."
import time

baseline_system_prompt = """
You are Albert Einstein. Follow these rules strictly:
- Answer directly — no preamble, no "Great question!", no restating the question
- Be brief: use the fewest words that fully answer the question
- No lengthy explanations unless explicitly asked
- No bullet points unless the answer is genuinely a list
- Stop as soon as the answer is complete
""".strip()

user_prompt = "Can you tell me about relativity?"

print(f"Sending prompt to {MODEL} model...")

t0 = time.perf_counter()
baseline_response = get_completion(baseline_system_prompt, user_prompt)  # your existing call
elapsed = round(time.perf_counter() - t0, 3)

#baseline_response = get_completion(baseline_system_prompt, user_prompt)
print("Response received!\n")

display_responses_modified(
    {
        "system_prompt": baseline_system_prompt,
        "user_prompt": user_prompt,
        "response": baseline_response,
    }
)
```

## 3. Define Persona-Specific Attributes (3 min)

Now, let's enhance the prompt by adding specific attributes about Einstein's personality, era-specific vocabulary, and expertise.

```python
# TODO: Add persona-specific attributes where you see **********
persona_system_prompt = f"""{baseline_system_prompt}.

Adopt these persona characteristics:

- Personality: **********
- Speech style: **********
- Expertise: **********
- Historical context: **********
  
Answer as if you are Einstein speaking in 1950, reflecting on your life and work.
Only discuss information that would have been known to you in your lifetime."""

user_prompt = "Can you tell me about relativity?"

print(f"Sending prompt to {MODEL} model...")
persona_response = get_completion(persona_system_prompt, user_prompt)
print("Response received!\n")

# Show last two prompts and responses
display_responses(
    {
        "system_prompt": baseline_system_prompt,
        "user_prompt": user_prompt,
        "response": baseline_response,
    },
    {
        "system_prompt": persona_system_prompt,
        "user_prompt": user_prompt,
        "response": persona_response,
    },
)
```
