# 📊 Ollama Response Statistics Dashboard
**Industry-standard LLM inference metrics for `qwen3:4b` running locally via Ollama**

Metrics tracked:
| Metric | Description |
|--------|-------------|
| **TTFT** | Time To First Token |
| **Total Latency** | Wall-clock time for full response |
| **Throughput** | Tokens/second (generation speed) |
| **Token counts** | Prompt, completion, total |
| **Tokens/min** | Useful for rate-limit planning |
| **Eval duration** | Pure model inference time |
| **Load time** | Model load time (cold start) |

```python
import time
import json
import requests
import ollama
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from openai import OpenAI
from dataclasses import dataclass, field, asdict
from typing import Optional
from IPython.display import display, HTML

# ── Config ────────────────────────────────────────────────────────
MODEL_NAME   = "qwen3:4b"
OLLAMA_HOST  = "http://localhost:11434"

SYSTEM_PROMPT = "You are a helpful, concise AI assistant. Keep answers short unless asked for detail."

print("✅ All imports successful")
print(f"   Model : {MODEL_NAME}")
print(f"   Host  : {OLLAMA_HOST}")
```

## 🧱 Step 1 — Stats Dataclass (Industry-Standard Metrics)

```python
@dataclass
class LLMStats:
    """
    Industry-standard inference metrics for a single LLM response.
    All times are in seconds unless noted.
    """
    # ── Identification ────────────────────────────────────────────
    model:              str   = ""
    method:             str   = ""   # ollama_sdk | rest_api | openai_compat
    prompt_preview:     str   = ""

    # ── Latency ───────────────────────────────────────────────────
    ttft_sec:           float = 0.0  # Time To First Token (streaming)
    total_latency_sec:  float = 0.0  # Wall-clock: request → last token
    load_duration_sec:  float = 0.0  # Time to load model into memory
    eval_duration_sec:  float = 0.0  # Pure inference time (no load)
    prompt_eval_sec:    float = 0.0  # Time spent processing prompt tokens

    # ── Token Counts ──────────────────────────────────────────────
    prompt_tokens:      int   = 0    # Input tokens
    completion_tokens:  int   = 0    # Output tokens generated
    total_tokens:       int   = 0    # prompt + completion

    # ── Throughput ────────────────────────────────────────────────
    tokens_per_second:  float = 0.0  # Generation speed (industry standard)
    tokens_per_minute:  float = 0.0  # tokens_per_second × 60
    prompt_eval_tps:    float = 0.0  # Prompt processing speed

    # ── Response ──────────────────────────────────────────────────
    response_text:      str   = ""
    response_chars:     int   = 0
    chars_per_second:   float = 0.0

    def compute_derived(self):
        """Auto-compute all derived metrics from raw values."""
        self.total_tokens    = self.prompt_tokens + self.completion_tokens
        self.response_chars  = len(self.response_text)

        if self.eval_duration_sec > 0:
            self.tokens_per_second = round(self.completion_tokens / self.eval_duration_sec, 2)
        elif self.total_latency_sec > 0:
            self.tokens_per_second = round(self.completion_tokens / self.total_latency_sec, 2)

        self.tokens_per_minute = round(self.tokens_per_second * 60, 2)

        if self.prompt_eval_sec > 0:
            self.prompt_eval_tps = round(self.prompt_tokens / self.prompt_eval_sec, 2)

        if self.total_latency_sec > 0:
            self.chars_per_second = round(self.response_chars / self.total_latency_sec, 2)

        return self

    def display(self):
        """Pretty-print all metrics in the notebook."""
        sep = "─" * 52
        print(f"\n{'═'*52}")
        print(f"  📊 LLM INFERENCE STATS — {self.method.upper()}")
        print(f"{'═'*52}")
        print(f"  Model   : {self.model}")
        print(f"  Prompt  : {self.prompt_preview[:45]}...")
        print(f"{sep}")

        print(f"  ⏱  LATENCY")
        print(f"     Total Response Time   : {self.total_latency_sec:.3f}s")
        if self.ttft_sec:
            print(f"     Time To First Token   : {self.ttft_sec:.3f}s  ← TTFT")
        print(f"     Model Load Time       : {self.load_duration_sec:.3f}s")
        print(f"     Pure Inference Time   : {self.eval_duration_sec:.3f}s")
        print(f"     Prompt Processing     : {self.prompt_eval_sec:.3f}s")
        print(f"{sep}")

        print(f"  🔢 TOKEN COUNTS")
        print(f"     Prompt Tokens         : {self.prompt_tokens}")
        print(f"     Completion Tokens     : {self.completion_tokens}")
        print(f"     Total Tokens          : {self.total_tokens}")
        print(f"{sep}")

        print(f"  🚀 THROUGHPUT")
        print(f"     Tokens / Second       : {self.tokens_per_second:.1f} tok/s")
        print(f"     Tokens / Minute       : {self.tokens_per_minute:.0f} tok/min")
        print(f"     Prompt Eval Speed     : {self.prompt_eval_tps:.1f} tok/s")
        print(f"     Chars / Second        : {self.chars_per_second:.1f} ch/s")
        print(f"{'═'*52}\n")

print("✅ LLMStats class defined")
```

---
## 🟢 Method 1 — ollama SDK with Full Stats

```python
def query_ollama_sdk(user_prompt: str, system_prompt: str = SYSTEM_PROMPT) -> LLMStats:
    """
    Query Ollama using the official Python SDK and collect full stats.
    Uses streaming to measure TTFT accurately.
    """
    stats = LLMStats(model=MODEL_NAME, method="ollama_sdk",
                     prompt_preview=user_prompt)

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user",   "content": user_prompt},
    ]

    t_start   = time.perf_counter()
    first_tok = True
    full_text = []

    stream = ollama.chat(model=MODEL_NAME, messages=messages, stream=True)

    for chunk in stream:
        if first_tok:
            stats.ttft_sec = round(time.perf_counter() - t_start, 4)
            first_tok = False

        tok = chunk["message"]["content"]
        full_text.append(tok)
        print(tok, end="", flush=True)   # live output

        # Final chunk carries timing metadata from Ollama
        if chunk.get("done"):
            ns = 1_000_000_000  # nanoseconds → seconds
            stats.load_duration_sec  = round(chunk.get("load_duration",   0) / ns, 4)
            stats.eval_duration_sec  = round(chunk.get("eval_duration",   0) / ns, 4)
            stats.prompt_eval_sec    = round(chunk.get("prompt_eval_duration", 0) / ns, 4)
            stats.prompt_tokens      = chunk.get("prompt_eval_count", 0)
            stats.completion_tokens  = chunk.get("eval_count", 0)

    stats.total_latency_sec = round(time.perf_counter() - t_start, 4)
    stats.response_text     = "".join(full_text)
    stats.compute_derived()

    print("\n")  # newline after streaming
    return stats


# ── Run it ────────────────────────────────────────────────────────
print("🤖 Response:\n")
stats1 = query_ollama_sdk("Explain how transformers work in deep learning.")
stats1.display()
```

---
## 🔵 Method 2 — REST API with Full Stats

```python
def query_rest_api(user_prompt: str, system_prompt: str = SYSTEM_PROMPT) -> LLMStats:
    """
    Query Ollama via raw REST API /api/chat and collect full stats.
    Streams NDJSON lines to capture TTFT and Ollama's native timing fields.
    """
    stats = LLMStats(model=MODEL_NAME, method="rest_api",
                     prompt_preview=user_prompt)

    payload = {
        "model": MODEL_NAME,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_prompt},
        ],
        "stream": True,
    }

    t_start   = time.perf_counter()
    first_tok = True
    full_text = []

    with requests.post(
        f"{OLLAMA_HOST}/api/chat",
        json=payload, stream=True, timeout=120
    ) as resp:
        resp.raise_for_status()

        for line in resp.iter_lines():
            if not line:
                continue
            chunk = json.loads(line)

            tok = chunk.get("message", {}).get("content", "")
            if tok:
                if first_tok:
                    stats.ttft_sec = round(time.perf_counter() - t_start, 4)
                    first_tok = False
                full_text.append(tok)
                print(tok, end="", flush=True)

            if chunk.get("done"):
                ns = 1_000_000_000
                stats.load_duration_sec  = round(chunk.get("load_duration",        0) / ns, 4)
                stats.eval_duration_sec  = round(chunk.get("eval_duration",        0) / ns, 4)
                stats.prompt_eval_sec    = round(chunk.get("prompt_eval_duration", 0) / ns, 4)
                stats.prompt_tokens      = chunk.get("prompt_eval_count", 0)
                stats.completion_tokens  = chunk.get("eval_count",        0)

    stats.total_latency_sec = round(time.perf_counter() - t_start, 4)
    stats.response_text     = "".join(full_text)
    stats.compute_derived()

    print("\n")
    return stats


# ── Run it ────────────────────────────────────────────────────────
print("🤖 Response:\n")
stats2 = query_rest_api("What are the SOLID principles in software engineering?")
stats2.display()
```

## 🟡 Method 3 — OpenAI-Compatible SDK with Stats

```python
def query_openai_compat(user_prompt: str, system_prompt: str = SYSTEM_PROMPT) -> LLMStats:
    """
    Query Ollama via OpenAI-compatible SDK and collect available stats.
    Note: TTFT is measured client-side; Ollama's native ns timings are
    not returned by the /v1/chat/completions endpoint.
    """
    stats = LLMStats(model=MODEL_NAME, method="openai_compat",
                     prompt_preview=user_prompt)

    client = OpenAI(
        base_url=f"{OLLAMA_HOST}/v1",
        api_key="ollama",
    )

    t_start   = time.perf_counter()
    first_tok = True
    full_text = []
    usage     = None

    stream = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_prompt},
        ],
        stream=True,
        stream_options={"include_usage": True},   # request token counts
    )

    for chunk in stream:
        delta = chunk.choices[0].delta.content if chunk.choices else None
        if delta:
            if first_tok:
                stats.ttft_sec = round(time.perf_counter() - t_start, 4)
                first_tok = False
            full_text.append(delta)
            print(delta, end="", flush=True)

        # Token usage arrives in the final chunk
        if chunk.usage:
            usage = chunk.usage

    stats.total_latency_sec = round(time.perf_counter() - t_start, 4)
    stats.response_text     = "".join(full_text)

    if usage:
        stats.prompt_tokens     = usage.prompt_tokens
        stats.completion_tokens = usage.completion_tokens

    stats.compute_derived()
    print("\n")
    return stats


# ── Run it ────────────────────────────────────────────────────────
print("🤖 Response:\n")
stats3 = query_openai_compat("Briefly explain the difference between SQL and NoSQL databases.")
stats3.display()
```

## 📈 Step 4 — Multi-Run Benchmark (Same Prompt, 3 Runs)

```python
BENCH_PROMPT = "List 5 best practices for REST API design."
N_RUNS = 3
bench_results = []

print(f"Running {N_RUNS} benchmark iterations with ollama SDK...\n")
for i in range(N_RUNS):
    print(f"── Run {i+1}/{N_RUNS} ──────────────────────────")
    s = query_ollama_sdk(BENCH_PROMPT)
    bench_results.append({
        "Run":              i + 1,
        "Total Latency (s)": s.total_latency_sec,
        "TTFT (s)":         s.ttft_sec,
        "Tokens/s":         s.tokens_per_second,
        "Tokens/min":       s.tokens_per_minute,
        "Completion Tokens": s.completion_tokens,
        "Prompt Tokens":    s.prompt_tokens,
        "Load Time (s)":    s.load_duration_sec,
        "Eval Time (s)":    s.eval_duration_sec,
    })

df = pd.DataFrame(bench_results).set_index("Run")

# Add summary row
df.loc["AVG"] = df.mean()
df.loc["MIN"] = df.min()
df.loc["MAX"] = df.max()

print("\n📊 Benchmark Results:\n")
display(df.style
    .format({
        "Total Latency (s)": "{:.3f}",
        "TTFT (s)":          "{:.3f}",
        "Tokens/s":          "{:.1f}",
        "Tokens/min":        "{:.0f}",
        "Load Time (s)":     "{:.3f}",
        "Eval Time (s)":     "{:.3f}",
    })
    .background_gradient(cmap="YlGn", subset=["Tokens/s", "Tokens/min"])
    .background_gradient(cmap="YlOrRd_r", subset=["Total Latency (s)", "TTFT (s)"])
)
```

## 📊 Step 5 — Visualise Stats (Charts)

```python
# ── Compare all 3 methods side-by-side ───────────────────────────
all_stats = [stats1, stats2, stats3]
methods   = [s.method for s in all_stats]

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle(f"Ollama {MODEL_NAME} — Inference Stats by Connection Method",
             fontsize=14, fontweight="bold", y=1.01)

colors = ["#4CAF50", "#2196F3", "#FF9800"]

def bar(ax, values, title, ylabel, color_list, fmt=".2f"):
    bars = ax.bar(methods, values, color=color_list, edgecolor="white", linewidth=1.2)
    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.set_ylabel(ylabel, fontsize=9)
    ax.set_xticks(range(len(methods)))
    ax.set_xticklabels(methods, rotation=12, fontsize=8)
    for b, v in zip(bars, values):
        ax.text(b.get_x() + b.get_width()/2, b.get_height() + max(values)*0.01,
                f"{v:{fmt}}", ha="center", va="bottom", fontsize=9, fontweight="bold")
    ax.set_ylim(0, max(values) * 1.25)
    ax.spines[["top", "right"]].set_visible(False)

bar(axes[0,0],
    [s.total_latency_sec  for s in all_stats],
    "⏱ Total Latency", "Seconds", colors)

bar(axes[0,1],
    [s.ttft_sec           for s in all_stats],
    "⚡ Time To First Token (TTFT)", "Seconds", colors)

bar(axes[0,2],
    [s.tokens_per_second  for s in all_stats],
    "🚀 Throughput (tok/s)", "Tokens/Second", colors)

bar(axes[1,0],
    [s.tokens_per_minute  for s in all_stats],
    "📈 Tokens Per Minute", "Tokens/Min", colors, fmt=".0f")

bar(axes[1,1],
    [s.completion_tokens  for s in all_stats],
    "🔢 Completion Tokens", "Count", colors, fmt=".0f")

bar(axes[1,2],
    [s.eval_duration_sec  for s in all_stats],
    "🧠 Pure Inference Time", "Seconds", colors)

plt.tight_layout()
plt.savefig("ollama_stats.png", dpi=150, bbox_inches="tight")
plt.show()
print("📁 Chart saved as ollama_stats.png")
```

## 🏎️ Step 6 — Speed Rating & Interpretation

```python
def speed_rating(tps: float) -> str:
    """
    Industry reference benchmarks for local LLM throughput.
    Based on typical CPU/GPU performance tiers.
    """
    if tps >= 60:   return "🟣 Blazing  (GPU high-end / quantized)"
    if tps >= 30:   return "🟢 Fast     (GPU mid-range / Apple M-series)"
    if tps >= 15:   return "🟡 Moderate (GPU entry / CPU fast)"
    if tps >= 5:    return "🟠 Slow     (CPU standard)"
    return               "🔴 Very Slow (CPU low-end / thermal throttle)"

def ttft_rating(ttft: float) -> str:
    """TTFT benchmarks — human perception thresholds."""
    if ttft < 0.1:  return "🟢 Excellent (imperceptible)"
    if ttft < 0.5:  return "🟢 Good      (feels instant)"
    if ttft < 1.5:  return "🟡 Acceptable"
    if ttft < 3.0:  return "🟠 Noticeable"
    return               "🔴 Slow      (model loading / cold start)"

print("=" * 58)
print("  🏁  PERFORMANCE INTERPRETATION")
print("=" * 58)

for s in all_stats:
    print(f"\n  [{s.method}]")
    print(f"   Speed   : {speed_rating(s.tokens_per_second)}")
    print(f"            {s.tokens_per_second:.1f} tok/s  |  {s.tokens_per_minute:.0f} tok/min")
    print(f"   TTFT    : {ttft_rating(s.ttft_sec)}")
    print(f"            {s.ttft_sec:.3f}s")
    print(f"   Latency : {s.total_latency_sec:.3f}s total")

print(f"\n{'─'*58}")
print("  📚  INDUSTRY REFERENCE BENCHMARKS")
print(f"{'─'*58}")
ref = [
    ("GPT-4o (OpenAI API)",       "~80–120 tok/s",  "<0.3s"),
    ("Claude 3.5 Sonnet (API)",   "~70–100 tok/s",  "<0.3s"),
    ("Local 7B (Apple M2 GPU)",   "~40–70 tok/s",   "<0.5s"),
    ("Local 7B (NVIDIA RTX 4090)","~100–150 tok/s", "<0.2s"),
    ("Local 4B CPU (this model)", "~5–25 tok/s",    "0.5–3s"),
]
print(f"  {'System':<35} {'Throughput':<18} TTFT")
for name, tps, ttft in ref:
    print(f"  {name:<35} {tps:<18} {ttft}")
print()
```

## 🔁 Step 7 — Reusable Helper (Drop-In for Any Script)

```python
def chat_with_stats(
    user_prompt:   str,
    system_prompt: str   = SYSTEM_PROMPT,
    model:         str   = MODEL_NAME,
    verbose:       bool  = True,
) -> LLMStats:
    """
    One-liner wrapper: send a prompt, get back response + full stats.
    Uses the ollama SDK (most accurate timing via native metadata).
    """
    stats = LLMStats(model=model, method="ollama_sdk",
                     prompt_preview=user_prompt)
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user",   "content": user_prompt},
    ]
    t_start, first_tok, buf = time.perf_counter(), True, []

    for chunk in ollama.chat(model=model, messages=messages, stream=True):
        tok = chunk["message"]["content"]
        if tok and first_tok:
            stats.ttft_sec = round(time.perf_counter() - t_start, 4)
            first_tok = False
        buf.append(tok)
        if verbose:
            print(tok, end="", flush=True)
        if chunk.get("done"):
            ns = 1_000_000_000
            stats.load_duration_sec = round(chunk.get("load_duration",        0) / ns, 4)
            stats.eval_duration_sec = round(chunk.get("eval_duration",        0) / ns, 4)
            stats.prompt_eval_sec   = round(chunk.get("prompt_eval_duration", 0) / ns, 4)
            stats.prompt_tokens     = chunk.get("prompt_eval_count", 0)
            stats.completion_tokens = chunk.get("eval_count",        0)

    stats.total_latency_sec = round(time.perf_counter() - t_start, 4)
    stats.response_text     = "".join(buf)
    stats.compute_derived()
    if verbose:
        print()
        stats.display()
    return stats


# ── Example usage ─────────────────────────────────────────────────
result = chat_with_stats("What is the CAP theorem in distributed systems?")

# Access any field programmatically
print(f"\n✅ Quick Access Example:")
print(f"   result.tokens_per_second  = {result.tokens_per_second}")
print(f"   result.tokens_per_minute  = {result.tokens_per_minute}")
print(f"   result.ttft_sec           = {result.ttft_sec}")
print(f"   result.total_latency_sec  = {result.total_latency_sec}")
print(f"   result.completion_tokens  = {result.completion_tokens}")
```
